```python
# importation des modules
import pandas as pd
import numpy as np
```

```python
# ouverture du fichier de travail
source = "HDR 2007-2008 Table 12.xlsx"
# read_excel est une fonction de pandas qui vous ouvrira un fichier Excel
# la "source" peut être n'importe où, soit dans un fichier local, soit dans une ressource en ligne distante
df = pd.read_excel(source)
# head imprimera les x premières lignes du fichier de données pour que nous puissions voir à quoi il ressemble
df.head(15)
```

```python
# Repositionner la ligne d'en-tête
df = pd.read_excel(source, header=1)
# https://stackoverflow.com/a/51794989
# Supprimez d'abord toutes les colonnes (axe=1) sans valeurs
df = df.dropna(how="all", axis=1)
# Supprimer toutes les lignes (axe=0) sans valeurs
df = df.dropna(how="all", axis=0).reset_index(drop=True)
df.head(15)
```

```python
# Nous pourrions simplement fixer les colonnes, mais obtenons une liste des informations qui sont déjà là  
# et utilisons cela pour guider les nouveaux noms que nous devons inclure.
columns = []
for c in df.columns:
    # Pour chaque colonne, obtenir les valeurs des 5 premières lignes
    # puis conservez toutes les chaînes (c'est-à-dire excluez-les si elles sont nulles) et ajoutez-les en colonne
    columns.append([str(v) for v in df[c][:5] if pd.notnull(v)])
columns
```

```python
columns = [
    "HDI rank",
    "HDI level"
    "Country",
    "Adult_literacy_rate 1985-1994 (%)",
    "Reference 1",
    "Adult_literacy_rate 1995-2005 (%)",
    "Reference 2",
    "MDG_Youth_literacy_rate  1985-1994  (%)",
    "Reference 3",
    "MDG_Youth_literacy_rate 1995-2005 (%)",
    "Reference 4",
    "MDG_Net primary_enrol_rate_1991 (%)",
    "Reference 5",
    "MDG_Net primary_enrol_rate_2005 (%)",
    "Reference 6",
    "Net second_ enrol_rate 1991 (%)",   
    "Reference 7",
    "Net second_ enrol_rate 2005 (%)",   
    "Reference 8",
    "MDG_Child_reach_grade 5 _1991 (%)",
    "Reference 9",
    "MDG_Child_reach_grade 5 _2004 (%) ",
    "Reference 10",
    "Tertiary_stud_rate_1999-2005 (%)",
    "Reference 11"
]
# Réinitialiser les colonnes des cadres de données
df.columns = columns
# Supprimer les lignes qui contenaient les informations dont nous n'avons plus besoin
# et réinitialiser l'index
# np.arange(5) crée un tableau = [0, 1, 2, 3, 4, 5]
df = df.drop(np.arange(6), axis=0).reset_index(drop=True)
df.head(20)
```

```python
# Ce fichier n'est pas accompagné d'un fichier de métadonnées distinct. Toutes les références et notes de bas de page  
# se trouvent littéralement au bas du fichier. Jetons un coup d'œil :
df.tail(30)
```

```python
# Déposer les lignes du bas des données que nous voulons conserver, jusqu'à la fin (défini comme len(df))
df = df.drop(np.arange(175,len(df)), axis=0).reset_index(drop=True)
df.tail(20)
```
